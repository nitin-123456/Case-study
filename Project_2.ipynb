### Importing modules and accessing file

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file = pd.read_csv('C:\\Users\\dell\\Downloads\\startup_funding.csv')
#file = pd.read_csv('startup_funding.csv')
```

#### Task-1 Locations receiving most amount of funding

```python
file.CityLocation.replace('Delhi','New Delhi',inplace = True)
file.CityLocation.replace('bangalore','Bangalore',inplace = True)
file.CityLocation.fillna('Unknown',inplace = True)
d= {}
np_loc = np.array(file.CityLocation,dtype = str)
for i in np_loc:
    if len(i.split('/')) == 2:
        j = i.split('/')
        for k in range(len(j)):
            j[k] = j[k].strip()
        d[j[0]] = d.get(j[0],0)+1
        #d[j[1]] = d.get(j[1],0)+1
        
    else:
        d[i] = d.get(i,0)+1
del d['Unknown']
region = []
funds = []
for i in d:
    if i == 'Mumbai':
        region.append(i)
        funds.append(d[i])
    if i == 'Bangalore':
        region.append(i)
        funds.append(d[i])
    if i == 'Noida':
        region.append(i)
        funds.append(d[i])
    if i == 'New Delhi':
        region.append(i)
        funds.append(d[i])
    if i == 'Gurgaon':
        region.append(i)
        funds.append(d[i])
ans = max(funds)
for i in range(len(region)):
    if funds[i] == ans:
        print(region[i])
plt.plot(region,funds)
plt.xlabel('Location')
plt.ylabel('No of Funding received')
plt.title('Funding vs Location')
plt.show()
```

### Task 2 - Top 5 Investors

```python
file.InvestorsName.fillna('Unknown',inplace = True)
d = {}
for i in file.InvestorsName:
    if len(i.split(','))>1:
        j = i.split(',')
        for k in j:
            d[k.strip()] = d.get(k.strip(),0)+1
    else:
        d[i] = d.get(i,0)+1
investors = []
times_invested = list(sorted(d.values()))
times_invested.reverse()
for i in range(5):
    for j in d:
        if times_invested[i] == d[j]:
            investors.append(j)
for i in range(len(investors)):
    print(investors[i],times_invested[i])
    
plt.pie(times_invested[0:5],labels = investors,autopct = '%.02f')
plt.axis('equal')
plt.show()
```

### Task-3 - Unique Investors

```python
df = file.copy()
df['StartupName'].replace('Flipkart.com','Flipkart',inplace=True)
df['StartupName'].replace('Ola Cabs','Ola',inplace=True)
df['StartupName'].replace('Olacabs','Ola',inplace=True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace=True)
df['StartupName'].replace('OyoRooms','Oyo',inplace=True)
df['StartupName'].replace('Oyorooms','Oyo',inplace=True)
df['StartupName'].replace('OYO Rooms','Oyo',inplace=True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True)
df.InvestorsName.fillna('Unknown',inplace = True)
d = df.set_index('StartupName').to_dict()['InvestorsName']
insp = {}# investors with unique startups
for i in d:
    if(len(d[i].split(','))>1):
        j = d[i].split(',')
        for k in j:
            if insp.get(k.strip()) == None:
                insp[k.strip()] = i,
            else:
                if i not in insp[k.strip()]:
                    insp[k.strip()] = insp[k.strip()]+(i,)
        
    else:
        if insp.get(d[i])== None:
            insp[d[i]] = i,
        else:
            if i not in insp[d[i]]:
                insp[d[i]] = insp[d[i]]+(i,)
    
del insp['']
del insp['Unknown']
for i in insp:
    insp[i] = len(insp[i])
a = list(sorted(insp.values()))
a.reverse()
ans = []
i = 0
while(i<5):
    for j in insp:
        if a[i] == insp[j]:
            ans.append(j)
            i = i+1
for i in ans[0:5]:
    print(i)
    pass
```
